```python
def count_number_of_unique_tokens(df):

    unique_tokens = set()
    for i in range(len(df)):
        tokens = indic_tokenize.trivial_tokenize(df.loc[i,'news_article'], lang = "ta")
        for token in tokens:
            unique_tokens.add(token)
    return sorted(unique_tokens), len(unique_tokens)
```
